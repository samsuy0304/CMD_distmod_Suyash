```python
import numpy as np
import matplotlib.pyplot as plt

#Constants
h = 6.63*(10**(-34))
c = 2.99*(10**8)
k_b= 1.380649*(10**(-23))
Ts = 5800
Te = 300
e=2.718

lam_low = 0.2*(10**(-6))
lam_high =20.0*(10**(-6))

# Converting Wavelenght to Frequency
def Lam_to_freq(lam):
    nu = c/lam
    return nu


def Radition(lam,T):
    nu = Lam_to_freq(lam)

    a = (2*h*(nu)**3)/(c**2)
    b = e**((h*nu)/(k_b*T))
    
    ce = (1)/(b-1)
    d  = a*ce
    return d




def Jean(lam,T ):
    nu = Lam_to_freq(lam)
    a = (2*k_b*T*(nu**2))/(c**2)
    return a
print(Radition(0.25*(10**(-6)),Ts))
x= []
ys = []
ye=[]
yjs=[]
yje=[]
ytest = []

rt =int((lam_high-lam_low)/(0.01*(10**(-6))))
for i in range(rt+1):
    x.append(lam_low)

    Bs = Radition(lam_low,Ts)
    Bts = Rad(lam_low,Ts)
    Be = Radition(lam_low,Te)
    Bjs = Jean(lam_low,Ts)
    Bje = Jean(lam_low,Te)

    ye.append(Be)
    ys.append(Bs)
    ytest.append(Bts)
    yje.append(Bje)
    yjs.append(Bjs)
    lam_low += 0.01*(10**(-6))

print(x)
plt.plot(x,ys,'r-', label = "Sun")
plt.plot(x,ye,'b-', label = "Earth" )
plt.plot(x,yjs,'g-', label = "Sun_Jean")
plt.plot(x,yje,'m-', label = "Earth_Jean" )
plt.yscale('log')
plt.gca().set_xlabel(r'$\lambda$ (meters)')
plt.gca().set_ylabel(r'log($B_{\lambda}$) ($W \cdot sr^{-1} \cdot m^3$)')
plt.ylim(0, 10**14)
plt.legend()

plt.show()   
    

```

```python
#Necessary Library
import numpy as np

#Setting Required Constants
h = 6.63*(10**(-34))
c = 2.99*(10**8)
k_b= 1.380649*(10**(-23))
Ts = 5800 # Sun temp
Te = 300  # Earth temp
e=2.718
rad_sun = 6.955*(10**5) #Sun Radius
rad_earth = 6.3*(10**3) # Earth Radius

pc = 3.086*(10**13)  # Conversion of parsecs to km
d_qb = 10*pc # 

def Rad(lam,T):
    a = (2*h*(c)**2)/(lam**5)
    b = e**((h*c)/(lam*k_b*T))
    
    ce = (1)/(b-1)
    d  = a*ce
    return d

def F_lam(lam,T,R,d):
    I = Rad(lam,T)

    F = np.pi*I*((R/d)**2)

    return F


#Range of Wavelength
lam_low = 0.2*(10**(-6))
lam_high =20.0*(10**(-6))



x= []
ys = []
ye=[]
ys10=[]
ye10=[]




rt =int((lam_high-lam_low)/(0.01*(10**(-6))))
for i in range(rt+1):
    x.append(lam_low)
    F_s = F_lam(lam_low,Ts,1,1)
    F_e = F_lam(lam_low,Te,1,1)
    F_s10 = F_lam(lam_low,Ts,rad_sun,d_qb)
    F_e10 = F_lam(lam_low,Te,rad_earth,d_qb)
    
    ye.append(F_e)
    ys.append(F_s)
    ye10.append(F_e10)
    ys10.append(F_s10)
    lam_low += 0.01*(10**(-6))




x = np.array(x)
ys = np.array(ys)
ye = np.array(ye)

Sun_Flux = np.sum(x*ys)
Earth_Flux = np.sum(x*ye)

print(Sun_Flux)
print(Earth_Flux)
print(Earth_Flux/Sun_Flux)
import matplotlib.pyplot as plt

plt.plot(x,ys,'r-', label = "Sun")
plt.plot(x,ye,'b-', label = "Earth" )

plt.gca().set_xlabel(r'$\lambda$ (meters)')
plt.gca().set_ylabel(r'$B_{\lambda}$ ($W \cdot m^{-3}$)')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import log10 , floor
def round_it(x, sig):
    return round(x, sig-int(floor(log10(abs(x))))-1)

M_Earth = 5.972*10**24 # Kilograms
R_Earth = 6.4*10**6 # meters
G= 6.67*(10**(-11))

#Find the constant density of Earth

V_Earth = (4/3)*np.pi*(R_Earth**3)
rho_0 = round_it(M_Earth/V_Earth, 6)

print("\n\nConstant density of Earth, rho_0 =", rho_0, "$kg/m^{3}\n\n")
```

```python
#Calculate and Plot the enclosed Mass 
r = np.arange(1,R_Earth,1) # Infinitesimal increments in radius

def Mass(r, u):
    m = u*(4/3)*np.pi*(r**3)
    return m

Vector_Mass = np.vectorize(Mass)

Enclosed_Mass = Vector_Mass(r,rho_0)


plt.plot(r,Enclosed_Mass, 'r-', label="Earth")
plt.axvline(x=R_Earth, ymin=0.05, ymax=0.95, color='b', label='Radius of Earth', ls='--')
plt.gca().set_xlabel(r'Radius')
plt.gca().set_ylabel(r'Mass')
plt.legend()
plt.show()


# Calculating Mass for r=R
print('\n\n Enclosed Mass of Earth =',round_it(Mass(R_Earth, rho_0),3), "kg")



# Comparison
#The enclosed mass for R=r seems to be equal to the true mass of earth.
```

```python
def gravity(m,r):
    g = G*(m/(r**2))
    return g

Vector_gravity = np.vectorize(gravity)

Gravity_field = Vector_gravity(Enclosed_Mass, r)

plt.plot(r,Gravity_field, 'r-', label="Earth")
plt.axvline(x=R_Earth, ymin=0.05, ymax=0.95, color='b', label='Radius of Earth', ls='--')
plt.gca().set_xlabel(r'Radius')
plt.gca().set_ylabel(r'Gravity')
plt.legend()
plt.show()

print("\n\n Surface gravity = ",gravity(M_Earth,R_Earth), "m/s^2")


# Comparison
#It seems that the surface gravity comes out to be lower than the actual
#value of 9.98 m/s^2
```

```python
def Pressure(a,b):
    c = -a*b
    return c

plt.plot(r,Pressure(rho_0,Gravity_field), 'r-', label="Earth")
plt.axvline(x=R_Earth, ymin=0.05, ymax=0.95, color='b', label='Radius of Earth', ls='--')
plt.gca().set_xlabel(r'Radius')
plt.gca().set_ylabel(r'Internal Pressure')
plt.legend()
plt.show()
```
